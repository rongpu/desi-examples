```python
from __future__ import division, print_function
import numpy as np
# import matplotlib
# matplotlib.use("Agg")
import matplotlib.pyplot as plt
from astropy.table import Table
import fitsio
```

```python
import desi_plot_spectrum
```

```python
coadd_fn = '/global/cfs/cdirs/desi/spectro/redux/fuji/tiles/cumulative/80605/20210205/coadd-0-80605-thru20210205.fits'
tid = 39627640566453451
```

__Observed-frame model spectrum__

```python
wave, model_flux = desi_plot_spectrum.get_rr_model(coadd_fn, tid, use_targetid=True)

plt.figure(figsize=(15, 7))
for camera in ['B', 'R', 'Z']:
    plt.plot(wave[camera], model_flux[camera], lw=1)
plt.xlabel('wavelength ($\AA$)')
plt.grid(alpha=0.5)
plt.show()
```

__Observed and model spectra__

```python
desi_plot_spectrum.plot_spectrum(coadd_fn, tid, use_targetid=True)
plt.show()
```

__Restframe model spectrum__

```python
wave, model_flux = desi_plot_spectrum.get_rr_model(coadd_fn, tid, use_targetid=True, restframe=True)

plt.figure(figsize=(15, 7))
plt.plot(wave, model_flux, lw=1)
plt.xlabel('wavelength ($\AA$)')
plt.grid(alpha=0.5)
plt.show()
```
